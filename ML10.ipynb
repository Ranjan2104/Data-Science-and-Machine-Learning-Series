# Filtering Datafram with AND & OR Operators

```python
import pandas as pd

df = pd.read_csv("employees.csv", parse_dates = ["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype(bool)
df["Gender"] = df["Gender"].astype("category")

df.head()
```

```python
mask1 = df["Gender"] == "Male"
mask2 = df["Team"] == "Marketing"

df[mask1 & mask2].head()
```

```python
mask3 = df["Senior Management"]
mask4 = df["Start Date"] < "1990-01-01"

df[mask3 | mask4].head()
```

```python
mask5 = df["First Name"] == "Robert"
mask6 = df["Team"] == "Client Services"
mask7 = df["Start Date"] > "2016-06-01"

df[(mask5 & mask6) | mask7].head()
```

# .isin() method

```python
import pandas as pd

df = pd.read_csv("employees.csv", parse_dates = ["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype(bool)
df["Gender"] = df["Gender"].astype("category")

df.head()
```

```python
mask1 = df["Team"] == "Legal"
mask2 = df["Team"] == "Sales"
mask3 = df["Team"] == "Product"

df[mask1 | mask2 | mask3].head()
```

# or by using .isin() methods

```python
mask = df["Team"].isin(["Legal","Sales","Product"])
df[mask].head()
```

```python
df[df["Team"].isin(["Legal","Sales","Product"])].head() # in one line easy to use.
```
