# .apply() method

```python
import pandas as pd

bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond.sort_index(inplace = True)

bond.head()
```

```python
def add_millions(number):
    return str(number) + " Millions!"

bond["Box Office"] = bond["Box Office"]. apply(add_millions)
bond.head()
```

```python
bond["Budget"] = bond["Budget"]. apply(add_millions)
bond.head()
```

# or

```python
columns= ["Box Office","Budget","Bond Actor Salary"]
for col in columns:
    bond[col] = bond[col].apply(add_millions)
bond.head()
```

# .apply() method Row wise

```python
def good_movie(row):
    actor = row[1]
    budget = row[4]
    
    if actor == "Pierce Brosnan":
        return "The Best"
    elif actor == "Roger Moore" or budget == 40:
        return "Enjoyable"
    else:
        return "Inhave no idea"

bond.apply(good_movie, axis = "columns").head()
```
