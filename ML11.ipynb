# .isnull() & .notnull() method

```python
import pandas as pd

df = pd.read_csv("employees.csv", parse_dates = ["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype(bool)
df["Gender"] = df["Gender"].astype("category")

df.head()
```

```python
mask = df["Team"].isnull()
df[mask].head()
```

```python
mask = df["First Name"].isnull()
df[mask].head()
```

```python
mask = df["Team"].notnull()
df[mask].head()
```

```python
mask = df["Gender"].notnull()
df[mask].head()
```

# .Between() methods

```python
import pandas as pd

df = pd.read_csv("employees.csv", parse_dates = ["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype(bool)
df["Gender"] = df["Gender"].astype("category")

df.head()
```

```python
mask = df["Salary"].between(60000,70000)
df[mask].head()
```

```python
mask = df["Bonus %"].between(2.0,5.0)
df[mask].head()
```

```python
mask = df["Start Date"].between("1991-01-01","1992-01-01")
df[mask].head()
```

```python
mask = df["Last Login Time"].between("8:30 AM","12:00 PM")
df[mask].head()
```
