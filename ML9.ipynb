# Rank Method

```python
import pandas as pd

nba = pd.read_csv("nba.csv").dropna(how = "all")
nba["Salary"] = nba["Salary"].fillna(0).astype(int)

nba.head()
```

```python
nba["Salary"].rank(ascending = False).astype(int)
```

```python
nba["Salary Rank"] = nba["Salary"].rank(ascending = False).astype(int)
```

```python
nba.head()
```

```python
nba.sort_values(by = "Salary", ascending = False).head()
```

# Filtering of DataSet

```python
import pandas as pd

df = pd.read_csv("employees.csv")
df.head()
```

```python
df.info()  # size of dataset is 692+ kb due to inarrange of dataset
```

```python
df["Start Date"].head()
```

```python
pd.to_datetime(df["Start Date"]).head()
```

```python
df["Start Date"] = pd.to_datetime(df["Start Date"])
df.head()
```

```python
df["Last Login Time"].head()
```

```python
pd.to_datetime(df["Last Login Time"])
```

```python
df["Last Login Time"] = pd.to_datetime(df["Last Login Time"])
df.head()
```

```python
df["Senior Management"].head()
```

```python
df["Senior Management"].astype(bool)
```

```python
df["Senior Management"] = df["Senior Management"].astype(bool)
df.head()
```

```python
df["Gender"].astype("category")
```

```python
df["Gender"] = df["Gender"].astype("category")
df.head()
```

```python
df.info() # now the memory of dataset is reduce to 49kb only as earlier it was 62kb+.
```
