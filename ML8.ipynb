# Drop Rows with null values

```python
import pandas as pd
nba = pd.read_csv("nba.csv")
nba.head()
```

```python
nba.tail()
```

```python
nba.dropna()
```

```python
nba.dropna(how = "any").head()
```

```python
nba.dropna(how = "all")
```

```python
nba.dropna(how = "all",inplace= True)
```

```python
nba.tail()
```

```python
nba.dropna(axis = 1).head()
```

```python
nba.dropna(subset = ["Salary"])
```

# .sort_values() in dataframe

```python
import pandas as pd
nba = pd.read_csv("nba.csv")
nba.head()
```

```python
nba.sort_values("Name").head()
```

```python
nba.sort_values("Name", ascending = False).head()
```

```python
nba.sort_values("Age", ascending = False).head()
```

```python
nba.sort_values("Salary",na_position='last').head()
```

```python
nba.sort_values("Salary",na_position='last').tail()
```

```python
nba.sort_values("Salary",na_position='first').head()
```

```python
nba.sort_values("Salary",na_position='first').tail()
```

# .sort_values() for multiple columns

```python
import pandas as pd
nba = pd.read_csv("nba.csv")
nba.head()
```

```python
nba.sort_values(["Team", "Name"]).head()
```

```python
nba.sort_values(["Name", "Team"]).head()
```

```python
nba.sort_values(["Name", "Team"], ascending = False).head()
```

```python
nba.sort_values(["Name", "Team"], ascending = [False, True]).head()
```

```python
nba.sort_values(["Name", "Team"], ascending = [False, True],inplace = True)
```

```python
nba.head()
```
