# .duplicated() method

```python
import pandas as pd

df = pd.read_csv("employees.csv", parse_dates = ["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype(bool)
df["Gender"] = df["Gender"].astype("category")

df.sort_values("First Name",inplace = True)

df.head()
```

```python
mask = df["First Name"].duplicated()
df[mask].head()
```

```python
mask1 = df["First Name"].duplicated(keep = 'last')
df[mask1].head()
```

```python
mask2 = df["First Name"].duplicated(keep = False)
df[mask2].head()
```

```python
mask3 = ~df["First Name"].duplicated(keep = False)
df[mask3].head()
```

# .drop_duplicates() method

```python
import pandas as pd

df = pd.read_csv("employees.csv", parse_dates = ["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype(bool)
df["Gender"] = df["Gender"].astype("category")

df.sort_values("First Name",inplace = True)

df.head()
```

```python
len(df)
```

```python
len(df.drop_duplicates())
```

```python
mask = df.drop_duplicates(subset = ["First Name"],keep = "first")
```

```python
len(mask)
```

```python
mask1 = df.drop_duplicates(subset = ["First Name","Team"],keep = "last")
mask1.head()
```

```python
df.drop_duplicates(subset = ["First Name","Team"],inplace = True)
df.head()
```
