```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3,4,5,6]
y = [1,2,3,4,5,6]

plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('y-axis')
plt.title('Linear Graph')

plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [2,4,5,2,3,9]

plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('y-axis')
plt.title('Undirect Graph')

plt.show()
```

```python
t = np.arange(0.0,2.0,0.01)
s = 1 + np.cos(2 * np.pi * t)

plt.plot(t,s)
plt.xlabel('Time(t)')
plt.ylabel('Vol(mA)')
plt.title('Cosine Graph')

plt.show()
```

```python
t = np.arange(0.0,2.0,0.01)
s = 1 + np.cos(2 * np.pi * t)

plt.plot(t,s,'--')
plt.xlabel('Time(t)')
plt.ylabel('Vol(mA)')
plt.title('Cosine Graph')

plt.show()
```

```python
t = np.arange(0.0,2.0,0.01)
s = 1 + np.cos(2 * np.pi * t)

plt.plot(t,s)
plt.grid()
plt.xlabel('Time(t)')
plt.ylabel('Vol(mA)')
plt.title('Cosine Graph')

plt.show()
```

```python
t = np.arange(0.0,2.0,0.01)
s = 1 + np.cos(2 * np.pi * t)

plt.plot(t,s,color = 'red')
plt.xlabel('Time(t)')
plt.ylabel('Vol(mA)')
plt.title('Cosine Graph')

plt.show()
```

```python
t = np.arange(0.0,2.0,0.01)
s = 1 + np.cos(2 * np.pi * t)

plt.plot(t,s,'*', color = 'green')
plt.grid()
plt.xlabel('Time(t)')
plt.ylabel('Vol(mA)')
plt.title('Cosine Graph')

plt.show()
```

```python
x1 = np.linspace(0.0,5.0)
x2 = np.linspace(0.0,2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2,1,1)
plt.plot(x1,y1,'--' ,color = 'red')
plt.xlabel('X-axis')
plt.ylabel('y-axis')
plt.title('Cos Subplot-1')

plt.show()
```

```python
x1 = np.linspace(0.0,5.0)
x2 = np.linspace(0.0,2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2,1,2)
plt.plot(x1,y1,'*' ,color = 'black')
plt.xlabel('X-axis')
plt.ylabel('y-axis')
plt.title('Cos Subplot-2')

plt.show()
```

# Bar Graph

```python
x = [1,2,3,4,5]
y = [10,24,36,20,5]

my_label = ['one', 'two', 'three', 'four', 'five']
plt.bar(x,y,tick_label = my_label, width = 0.7, color = ['green','red','black','blue','pink'])

plt.xlabel("x-axis")
plt.ylabel('y-axis')
plt.title('Bar Graph')
plt.grid()

plt.show()
```

```python
x = [1,2,3,4,5]
y = [10,24,36,40,5]

my_label = ['one', 'two', 'three', 'four', 'five']
plt.bar(x,y,tick_label = my_label, width = 0.7, color = ['green','red','black','blue','pink'])

plt.xlabel("x-axis")
plt.ylabel('y-axis')
plt.title('Bar Graph')

plt.show()
```

# HistoGram plot

```python
# histogram plot
import matplotlib.pyplot as plt
ages = [2,50,70,40,30,45,50,45,43,40,44,60,7,13,51,18,90,77,32,21,20,40]

range = (0,100)
bins = 10  # gap

plt.hist(ages,bins,range,color='green',histtype = 'bar',rwidth = 0.7)
plt.xlabel('ages')
plt.ylabel('Bins')
plt.title('Histogram bar plot')
plt.show()
```
