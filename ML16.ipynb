# Create random sample with .sample() method

```python
import pandas as pd

bond = pd.read_csv("jamesbond.csv",index_col = "Film")
bond.sort_index(inplace = True)

bond.head()
```

```python
bond.sample()
```

```python
bond.sample(n = 5) # random sample generates
```

```python
bond.sample(frac = .25) # 25% data extract
```

```python
bond.sample(n = 3, axis ="rows")
```

```python
bond.sample(n = 3, axis ="index")
```

```python
bond.sample(n = 3, axis ="columns").head()
```

```python
bond.sample(n = 3, axis =1).head()
```

# The .nsmallest() & .nlargest() methods

```python
import pandas as pd

bond = pd.read_csv("jamesbond.csv",index_col = "Film")
bond.head()
```

```python
bond.sort_values("Box Office",ascending = False)
bond.head()
```

# Or

```python
bond.nlargest(n = 3, columns="Budget")
```

```python
    bond.nlargest(n = 3, columns="Box Office")
```

```python
bond.nsmallest(n = 3, columns="Box Office")
```

```python
bond.nsmallest(n = 3, columns="Bond Actor Salary")
```

```python
bond["Box Office"].nlargest(n = 5)
```

```python
bond["Year"].nsmallest(n = 3)
```
